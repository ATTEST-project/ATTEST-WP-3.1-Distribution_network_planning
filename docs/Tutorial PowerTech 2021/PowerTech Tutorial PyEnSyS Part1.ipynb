Now let's check that PyEnSyS runs! I will walk you through the code and results after this step.

```python
from pyensys.engines.main import pyeneConfig, pyeneClass
from pyomo.core import ConcreteModel
import pyomo.environ
from pyomo.opt import SolverFactory

conf = pyeneConfig()

conf.EM.settings['File'] = '/content/ResolutionTreeWeek.json'
conf.NM.settings['File'] = '/content/case4m1.json'

conf.NM.settings['Flag'] = True
conf.NM.settings['Losses'] = False
conf.NM.settings['Feasibility'] = True
conf.NM.settings['NoTime'] = 24  # Number of time steps
conf.NM.scenarios['Weights'] = [1 for _ in range(24)]  # Add weights 
    # to the time steps

conf.HM.settings['Flag'] = False    
# Scenarios
conf.NM.scenarios['NoDem'] = 2  # Number of demand profiles

conf.NM.hydropower['Number'] = 2  # Number of hydropower plants
conf.NM.hydropower['Bus'] = [1, 2]  # Location (bus) of hydro
conf.NM.hydropower['Max'] = [700, 400]  # Generation capacity
conf.NM.hydropower['Cost'] = [1, 1]  # Costs

# Create object
EN = pyeneClass(conf.EN)
# Initialise with selected configuration
EN.initialise(conf)

# DEMAND PROFILES
Dem1 = [0.4819, 0.5310, 0.5014, 0.4587, 0.4280, 0.4282, 0.5326, 0.7258, 
0.8000, 0.8058, 0.7997, 0.7930, 0.7980, 0.7516, 0.7371, 0.7653, 
0.9142, 1.0000, 0.9401, 0.8702, 0.7960, 0.7441, 0.6415, 0.4969]

Dem2 = [0.5272, 0.5521, 0.5076, 0.4580, 0.4224, 0.4059, 0.4369, 0.4927, 
0.5804, 0.6698, 0.7147, 0.7312, 0.7464, 0.6920, 0.6365, 0.6475, 
0.7496, 0.8689, 0.8854, 0.8421, 0.7675, 0.7148, 0.6274, 0.5115]

EN.set_Demand(1, Dem1)
EN.set_Demand(2, Dem2)

EN.set_Hydro(1, 800)
EN.set_Hydro(2, 2000)

# Run integrated pyene
m = EN.run(solver="pyomo")

EN.Print_ENSim(m)
```

## CODE EXPLANATION

We need to initially load the libraries that we will use to run PyEnSyS. We do that by using the following code:

```python
from pyensys.engines.main import pyeneConfig, pyeneClass
from pyomo.core import ConcreteModel
import pyomo.environ
from pyomo.opt import SolverFactory
```

Then after PyEnSyS loads a predefined set of parameters to run properly:

```python
conf = pyeneConfig()
```

The next code indicate the full path to the json files that contain the network to be simulated.

```python
conf.EM.settings['File'] = '/content/ResolutionTreeWeek.json'
conf.NM.settings['File'] = '/content/case4m1.json'
```

If you want to run an optimal power flow you need to define the following flag in "True". However, if you want to run a single node representation of the system you need to set the flag in "False".

```python
conf.NM.settings['Flag'] = True
```

PyEnSyS can include non-linear losses in the DC Optimal Power Flow setting the following flag in "True":

```python
conf.NM.settings['Losses'] = False
```

The user can decide to use the feasibility option in case the problem does not converge because of lack of generation and/or transmission capacity in the DC OPF. This can be done setting the following flag in "True":

```python
conf.NM.settings['Feasibility'] = True
```

The following 2 lines define the number of time steps in each representative day and the weight for each hour. In this example we have 24 time steps representing a whole day and the weight of each time step is 1 hour.

```python
conf.NM.settings['NoTime'] = 24  # Number of time steps
conf.NM.scenarios['Weights'] = [1 for _ in range(24)]  # Add weights 
    # to the time steps
```

The number of demand profiles or representative days are defined with the following option:

```python
# Scenarios
conf.NM.scenarios['NoDem'] = 2  # Number of demand profiles
```

The following 4 lines of code define how to define hydropower generators in the OPF. The number of hydropower generators is defined first. In the second line the location of those generators is defined. The maximum generation capacity is defined in line 3 for all generators. Finally, the cost of generation for each generator is defined in line 4.

```python
conf.NM.hydropower['Number'] = 2  # Number of hydropower plants
conf.NM.hydropower['Bus'] = [1, 2]  # Location (bus) of hydro
conf.NM.hydropower['Max'] = [700, 400]  # Generation capacity
conf.NM.hydropower['Cost'] = [1, 1]  # Costs
```

PyEnSyS is the initialised with the data defined above using the following commands:

```python
# Create object
EN = pyeneClass(conf.EN)
# Initialise with selected configuration
EN.initialise(conf)
```

The demand profiles (for representative days) are then defined using the following commands:

```python
# DEMAND PROFILES
Dem1 = [0.4819, 0.5310, 0.5014, 0.4587, 0.4280, 0.4282, 0.5326, 0.7258, 
0.8000, 0.8058, 0.7997, 0.7930, 0.7980, 0.7516, 0.7371, 0.7653, 
0.9142, 1.0000, 0.9401, 0.8702, 0.7960, 0.7441, 0.6415, 0.4969]

Dem2 = [0.5272, 0.5521, 0.5076, 0.4580, 0.4224, 0.4059, 0.4369, 0.4927, 
0.5804, 0.6698, 0.7147, 0.7312, 0.7464, 0.6920, 0.6365, 0.6475, 
0.7496, 0.8689, 0.8854, 0.8421, 0.7675, 0.7148, 0.6274, 0.5115]

EN.set_Demand(1, Dem1)
EN.set_Demand(2, Dem2)
```

The numbers 1 and 2 do not represent nodes but the number of the representative day or scenario.

The following commands are used to set one week equivalent of energy for each hydropower generator. The numbers in the set function do not represent the bus number but the position in the list for each generator.

```python
EN.set_Hydro(1, 800)
EN.set_Hydro(2, 2000)
```

Finally, the following commands start the simulation and print the results on the screen.

```python
# Run integrated pyene
m = EN.run(solver="pyomo")

EN.Print_ENSim(m)
```

If we convert the results in tables it will look like the tables below. Each vector correspond to one hydropower generator defined above. This mean that the optimisation algorithm uses all the water available during the weekdays and it runs out of water during the weekends. This is reflected in the operation of the hydropower generator 1 that will be presented later.
